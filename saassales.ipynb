# Explorating Data Analysis (EDA)

<div class="alert alert-info">

Exploratory Data Analysis (EDA) merupakan tahap kritis dalam analisis data karena memungkinkan analis untuk mengungkap pola, anomali, dan struktur yang tersembunyi dalam dataset. Melalui EDA, analis dapat memvisualisasikan distribusi variabel, mengidentifikasi outlier dan missing values, serta menemukan korelasi dan hubungan antar variabel yang mungkin tidak terlihat pada pandangan awal. Proses ini membantu dalam formulasi hipotesis yang tepat dan pemilihan model analisis yang sesuai dengan karakteristik data. EDA juga memfasilitasi pemahaman mendalam tentang dataset, sehingga memungkinkan transformasi data yang lebih efektif dan interpretasi hasil yang lebih akurat. 

## Uji Normalitas Numerik

Uji normalitas numerik adalah metode statistik yang digunakan untuk memeriksa apakah distribusi data numerik (seperti Sales, Profit, atau Quantity dalam dataset) mengikuti distribusi normal

```python
from scipy.stats import shapiro, kstest, normaltest
from statsmodels.stats.diagnostic import lilliefors

numeric = df[["Sales", "Quantity", "Discount", "Profit"]]
results = []

print("H0 = Data terdistribusi Normal")
print("H1 = Data tidak terdistribusi Normal")

for col in numeric.columns:
    ks_stat, ks_pval = kstest(df[col], 'norm')
    shapiro_stat, shapiro_pval = shapiro(df[col])
    lilliefors_stat, lilliefors_pval = lilliefors(df[col])
    normaltest_stat, normaltest_pval = normaltest(df[col])

    # Test results
    results.append({
        'Column': col,
        'KS_p-value': ks_pval,
        'Shapiro_p-value': shapiro_pval,
        'Pearson_p-value': normaltest_pval,
        'Lilliefors_p-value': lilliefors_pval,
        'KS_Result': 'Reject H0' if ks_pval < 0.05 else 'Tidak terbukti reject H0',
        'Shapiro_Result': 'Reject H0' if shapiro_pval < 0.05 else 'Tidak terbukti reject H0',
        'Lilliefors_Result': 'Reject H0' if lilliefors_pval < 0.05 else 'Tidak terbukti reject H0',
        'Pearson_Result': 'Reject H0' if normaltest_pval < 0.05 else 'Tidak terbukti reject H0'
    })

# Result to dataframe
results_df = pd.DataFrame(results)
results_df
```

<div class="alert alert-info">

Hasilnya, pada dala Sales, Quantity, Discount, dan Profit tidak ada yang terdistribusi normal.

## Outliers

### Boxplot

```python
# Boxplot

plt.figure(figsize = (15,10)) # kolom, baris

for posisi, nama_kolom in zip(range(1,10), numerik.columns): # posisi 1 sampai 9
    plt.subplot(3,3,posisi) # jumlah baris, jumlah kolom, posisi
    sns.boxplot(x = df[nama_kolom])

plt.show()
```

Kotak diagram memperlihatkan outlier yang signifikan pada Sales dan Profit, yang mengindikasikan variabilitas yang tinggi, sementara Kuantitas dan Discount memperlihatkan distribusi yang lebih konsisten dengan lebih sedikit outlier.

### IQR 1.5

```python
## IQR 1.5

for nama_kolom in numerik.columns:
    q3 = df[nama_kolom].quantile(0.75)
    q1 = df[nama_kolom].quantile(0.25)
    iqr = q3 - q1
    batas_atas = q3 + (iqr * 1.5)
    batas_bawah = q1 - (iqr * 1.5)
    jumlah_outliers = len(df[(df[nama_kolom] > batas_atas) | (df[nama_kolom] < batas_bawah)])
    print(f'Jumlah Outliers {nama_kolom}: {jumlah_outliers}')
```

### Z Score

```python
## Z Score

for kolom in numerik.columns:
    std = df[kolom].std()
    mean = df[kolom].mean()

    # menghitung zscore
    zscore = abs((df[kolom] - mean) / std)

    # mencari outliers
    outliers = len(df[zscore > 2])
    print(f'Jumlah outliers {kolom} : {outliers}')
```

# Korelasi

Uji korelasi adalah metode statistik yang digunakan untuk mengukur hubungan antara dua variabel numerik, termasuk kekuatan dan arah hubungan tersebut.

```python
numeric = df[["Sales", "Quantity", "Discount", "Profit"]]
sns.heatmap(numeric.corr(method="spearman"),annot=True)
plt.title("Spearman Correlation Heatmap")
plt.show()
```

Spearman method is used because the data is not normally distributed.

Heatmaps menunjukkan bahwa meskipun ada beberapa hubungan antara variabel-variabel tersebut, 
Yang paling signifikan adalah antara Sales dan Profit, dan antara Discount dan Profit (dalam cara yang negatif).

Hal ini menunjukkan bahwa strategi untuk meningkatkan Sales dapat berdampak positif pada Profit, sementara Discount yang berlebihan dapat berdampak negatif pada profitabilitas.

## Analisis (Industri)

### 3.1 Total penjualan per industri

```python
# 3.1 Total penjualan per industri
industry_sales_profit = df.groupby(['Industry', 'Year']).agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Profit_Ratio=('Profit', lambda x: x.sum() / df.loc[x.index, 'Sales'].sum()),  # Calculate Profit Ratio
    Total_Quantity=('Quantity', 'sum'),
    Total_Product=('Product', lambda x: x.value_counts().count())
).reset_index()

# Display the resulting DataFrame
industry_sales_profit
```

```python
total_sales_by_industry = industry_sales_profit.groupby('Industry')['Total_Sales'].sum().reset_index()

# Sort the data for better visualization
total_sales_by_industry = total_sales_by_industry.sort_values('Total_Sales', ascending=False)
unique_industries = total_sales_by_industry['Industry'].unique()

# Create the bar plot for total sales by industry
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Total_Sales', y='Industry', hue='Industry', data=total_sales_by_industry)

# Add labels to each bar, excluding zero values
for p in bar_plot.patches:
    width = p.get_width()
    if width > 0:  # Only annotate if width is greater than 0
        bar_plot.annotate(f'{width:.2f}', 
                          (width, p.get_y() + p.get_height() / 2.), 
                          ha='left', va='center', 
                          fontsize=10, color='black', 
                          xytext=(3, 0),  # Offset the label slightly to the right of the bar
                          textcoords='offset points')

# Set the title and labels for the plot
plt.title('Total Sales by Industry (2020-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales (USD)')
plt.ylabel('Industry')
plt.xlim(0, total_sales_by_industry['Total_Sales'].max() * 1.1)  # Adjust xlim for better spacing

plt.tight_layout()
plt.savefig('p_per_industri.png')
plt.show()
```

### 3.3 Margin profit per industri (Profit/Sales)

```python
# 3.3 Margin profit per industri (Profit/Sales)

print("\n3.3 MARGIN PROFIT PER INDUSTRI")
print("="*50)
industry_margin = df.groupby('Industry').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()
industry_margin['Persentase_Margin'] = ((industry_margin['Profit'] / industry_margin['Sales']) * 100).round(2)
industry_margin = industry_margin.sort_values('Persentase_Margin', ascending=False)
print(industry_margin[['Industry', 'Persentase_Margin']])

margin_profit_by_industry = industry_margin.groupby('Industry')['Persentase_Margin'].sum().reset_index()
# Sort the data for better visualization
margin_profit_by_industry = margin_profit_by_industry.sort_values('Persentase_Margin', ascending=False)
unique_industries = margin_profit_by_industry['Industry'].unique()

# Create the bar plot for total sales by industry
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Persentase_Margin', y='Industry', hue='Industry', data=margin_profit_by_industry)

# Add labels to each bar, excluding zero values
for p in bar_plot.patches:
    width = p.get_width()
    if width > 0:  # Only annotate if width is greater than 0
        bar_plot.annotate(f'{width:.2f}', 
                          (width, p.get_y() + p.get_height() / 2.), 
                          ha='left', va='center', 
                          fontsize=10, color='black', 
                          xytext=(3, 0),  # Offset the label slightly to the right of the bar
                          textcoords='offset points')

# Set the title and labels for the plot
plt.title('Margin Profit by Industry (2020-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Margin Profit (USD)')
plt.ylabel('Industry')
plt.xlim(0, margin_profit_by_industry['Persentase_Margin'].max() * 1.1)  # Adjust xlim for better spacing

plt.tight_layout()
plt.savefig('margin_per_industri.png')
plt.show()
```

## Analisis Segmen (Pelanggan)

### 5.1 Performa Segman Pelanggan

```python
print("\n5. PERFORMA SEGMEN PELANGGAN")
print("="*50)
segment_perf = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'  # Jumlah order unik
}).reset_index()
segment_perf['Profit_Margin'] = (segment_perf['Profit'] / segment_perf['Sales'] * 100).round(2)
segment_perf = segment_perf.sort_values('Sales', ascending=False)
segment_perf = segment_perf.rename(columns={'Order ID': 'Orders'})
print(segment_perf)

# Visualisasi performa segmen
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment', y='Sales', data=segment_perf)
plt.title('Penjualan per Segmen Pelanggan')
plt.tight_layout()
plt.savefig('penjualan_per_segmen.png')
plt.show()
```

<div class="alert alert-info">

- SMB (Small and Medium Business) adalah segmen dengan kontribusi penjualan tertinggi, mencapai hampir 1,2 juta unit/nominal. Ini menunjukkan bahwa bisnis skala kecil-menengah merupakan pasar utama dan paling potensial saat ini.

- Segmen Strategic menempati posisi kedua dengan total penjualan sekitar 700 ribu. Meskipun berada di bawah SMB, kontribusinya masih signifikan, menandakan adanya potensi pertumbuhan lebih lanjut terutama jika dikelola secara relasional dan berbasis solusi.

- Segmen Enterprise memiliki kontribusi paling rendah, dengan total penjualan sekitar 430 ribu. Ini mungkin disebabkan oleh siklus penjualan yang lebih panjang, keputusan pembelian yang lebih kompleks, atau kebutuhan akan penyesuaian produk yang tinggi.
